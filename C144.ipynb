## 1.4.4 Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('KakaoTalk_20210923_131053275.png')
n = 17
kernel = np.ones((n,n),np.float32)/(n*n)
dst = cv2.filter2D(img,-1,kernel)
plt.rcParams["figure.figsize"] = (20,10)
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img = cv2.imread('e5d9b3eb2fbfc27ecbeca4e238c85c86.jpg')
n = 11

blur = cv2.blur(img,(5,5))
gaussian = cv2.GaussianBlur(img,(n,n),0)
median = cv2.medianBlur(img,5)
bilateral = cv2.bilateralFilter(img,9,75,75)
plt.rcParams["figure.figsize"] = (20,10)

plt.subplot(221),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(blur),plt.title('Blur')
plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(gaussian),plt.title('GaussianBlur')
plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(median),plt.title('medianBlur')
plt.xticks([]), plt.yticks([])
plt.show()
```
